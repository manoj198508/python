### Q1. Write 5 names to `names.txt`, then read with line numbers
**Concept:** `'w'` writes (you add `\n` yourself); iterating the file with `enumerate(f, 1)` numbers the lines lazily.

**Explain to students:** `with` closes the file automatically; `line.strip()` removes the trailing newline when printing.

```python
names = ['Asha', 'Ravi', 'Mira', 'Neel', 'Sara']
with open('names.txt', 'w', encoding='utf-8') as f:
    for name in names:
        f.write(name + '\n')

with open('names.txt', encoding='utf-8') as f:
    for i, line in enumerate(f, 1):
        print(i, line.strip())
```

### Q2. Append a timestamped line to a log file on each run
**Concept:** mode `'a'` adds to the end without erasing existing content.

**Explain to students:** compare with `'w'` (which would wipe the log every time). Each call leaves a permanent trail.

```python
from datetime import datetime
def do_work():
    with open('log.txt', 'a', encoding='utf-8') as f:
        f.write(f'{datetime.now():%Y-%m-%d %H:%M:%S} - do_work ran\n')

do_work(); do_work()
print(Path('log.txt').read_text(encoding='utf-8'))
```

### Q3. Read a CSV of students and print the average CGPA
**Concept:** `csv.DictReader` maps each row to a dict using the header; values arrive as **strings**, so convert with `float()`.

**Explain to students:** we first create a small CSV, then read it back — mirroring a real data file.

```python
import csv
# create a sample file
with open('students.csv', 'w', newline='', encoding='utf-8') as f:
    w = csv.writer(f); w.writerow(['name', 'cgpa'])
    for row in [['Asha', 8.4], ['Ravi', 7.9], ['Mira', 9.1]]:
        w.writerow(row)

with open('students.csv', newline='', encoding='utf-8') as f:
    rows = list(csv.DictReader(f))
avg = sum(float(r['cgpa']) for r in rows) / len(rows)
print(f'Average CGPA over {len(rows)} students: {avg:.2f}')
```

### Q4. Save a settings dict to JSON, load it back, modify one key
**Concept:** `json.dump` writes a Python object as JSON; `json.load` reads it back into a dict — types (bool, list) are preserved.

**Explain to students:** JSON is how config files and web APIs exchange structured data.

```python
import json
settings = {'theme': 'dark', 'volume': 7, 'notify': True}
with open('config.json', 'w', encoding='utf-8') as f:
    json.dump(settings, f, indent=2)

with open('config.json', encoding='utf-8') as f:
    loaded = json.load(f)
loaded['volume'] = 10                     # modify one key
with open('config.json', 'w', encoding='utf-8') as f:
    json.dump(loaded, f, indent=2)
print('updated settings:', loaded)
```

### Q5. Count lines and words in a text file (memory-friendly)
**Concept:** iterate the file line by line (never loading it all) and tally as you go.

**Explain to students:** `line.split()` breaks a line into words; this single pass works even on huge files.

```python
lines = words = 0
with open('names.txt', encoding='utf-8') as f:
    for line in f:
        lines += 1
        words += len(line.split())
print(f'lines = {lines}, words = {words}')
```

### Q6. Open a non-existent file and handle `FileNotFoundError`
**Concept:** catch the *specific* exception so a missing file doesn't crash the whole program.

**Explain to students:** always anticipate that files may be missing, moved, or unreadable — handle it and continue or fall back to defaults.

```python
try:
    with open('does_not_exist.txt', encoding='utf-8') as f:
        print(f.read())
except FileNotFoundError as e:
    print('Handled gracefully ->', e)
```
