## 1. Pick the right data structure

```python
nums = [3, 1, 2, 3, 1]
print('list  :', nums)
print('set   :', set(nums), '-> unique, fast membership')
print('tuple :', tuple(nums), '-> fixed, hashable')
marks = {'Asha': 8.4, 'Ravi': 7.9}
print('dict  :', marks, '-> key lookup:', marks['Asha'])
print('3 in set?', 3 in set(nums))   # O(1) vs scanning a list
```

## 2. Comprehensions

```python
evens = [n for n in range(10) if n % 2 == 0]
squares = {n: n*n for n in range(1, 6)}
vowels = {c for c in 'mississippi' if c in 'aeiou'}
print('evens  :', evens)
print('squares:', squares)
print('vowels :', vowels)
```

## 3. Flexible functions: *args, **kwargs, lambda, higher-order

```python
def total(*args):
    return sum(args)
def describe(**kwargs):
    return ', '.join(f'{k}={v}' for k, v in kwargs.items())

print('total(1,2,3,4):', total(1, 2, 3, 4))
print('describe(...):', describe(debug=True, level=5))

people = [('Asha', 8.4), ('Ravi', 7.9), ('Mira', 9.1)]
print('sorted by cgpa desc:', sorted(people, key=lambda p: p[1], reverse=True))
print('map/filter:', list(map(lambda x: x*x, filter(lambda x: x > 1, [0, 1, 2, 3]))))
```

## 4. Generators — lazy and memory-friendly

```python
def fib():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

import itertools
print('first 10 fibs:', list(itertools.islice(fib(), 10)))

# sum of squares up to 1,000,000 without building a list
print('sum of squares:', sum(i*i for i in range(1_000_000)))
```

## 5. Error handling

```python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return 'undefined (division by zero)'
    finally:
        pass  # cleanup would go here

print(safe_divide(10, 2))
print(safe_divide(10, 0))

for raw in ['42', 'oops']:
    try:
        print('parsed', int(raw))
    except ValueError as e:
        print('cannot parse', repr(raw), '->', e)
```

## 6. Standard-library power tools

```python
from collections import Counter, defaultdict
import itertools, functools

text = 'the cat sat on the mat the cat'
print('word counts:', Counter(text.split()).most_common(2))

groups = defaultdict(list)
for w in text.split():
    groups[len(w)].append(w)
print('grouped by length:', dict(groups))

print('running products:', list(itertools.accumulate([1,2,3,4], func=lambda a,b:a*b)))
print('factorial(6) via reduce:', functools.reduce(lambda a,b:a*b, range(1,7)))
```

## 7. Worked problems
**FizzBuzz**, **word frequency**, and **two-sum** (using a dict for O(n)).

```python
# FizzBuzz 1..15
for n in range(1, 16):
    print('Fizz'*(n%3==0) + 'Buzz'*(n%5==0) or n, end=' ')
print()
```

```python
# Two-sum: indices of the two numbers adding to target — O(n) with a dict
def two_sum(nums, target):
    seen = {}
    for i, v in enumerate(nums):
        if target - v in seen:
            return (seen[target - v], i)
        seen[v] = i
    return None

print('two_sum([2,7,11,15], 9):', two_sum([2, 7, 11, 15], 9))
print('two_sum([3,2,4], 6)   :', two_sum([3, 2, 4], 6))
```

## 8. Sanity-check with asserts (mini test)

```python
def is_palindrome(s):
    s = ''.join(ch.lower() for ch in s if ch.isalnum())
    return s == s[::-1]

assert is_palindrome('Madam')
assert is_palindrome('A man, a plan, a canal: Panama')
assert not is_palindrome('python')
print('all palindrome tests passed')
```
