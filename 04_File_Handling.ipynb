# Module 4 — File Handling in Python
Make data persist on disk. We'll write/read text, use `with`, `pathlib`, and the `csv` & `json` modules. All files are created inside a local **`fh_demo/`** folder so this notebook is self-contained.

```python
import os
from pathlib import Path
Path('fh_demo').mkdir(exist_ok=True)
os.chdir('fh_demo')
print('working directory:', Path.cwd().name)
```

## 1. Write text with `with` (auto-closes the file)

```python
rows = ['Asha,8.4', 'Ravi,7.9', 'Mira,9.1']
with open('marks.txt', 'w', encoding='utf-8') as f:
    for r in rows:
        f.write(r + '\n')        # write() does NOT add a newline
print('wrote', len(rows), 'lines to marks.txt')
```

## 2. Read it back — four ways

```python
with open('marks.txt', encoding='utf-8') as f:
    print('--- read() whole file ---')
    print(f.read())

with open('marks.txt', encoding='utf-8') as f:
    print('--- iterate line by line (best for big files) ---')
    for i, line in enumerate(f, 1):
        print(i, line.rstrip())
```

## 3. Append a new record (mode `'a'`)

```python
with open('marks.txt', 'a', encoding='utf-8') as f:
    f.write('Neel,8.8\n')
print('after append:')
print(Path('marks.txt').read_text(encoding='utf-8'))
```

## 4. Paths the portable way with `pathlib`

```python
p = Path('marks.txt')
print('name      :', p.name)
print('suffix    :', p.suffix)
print('exists    :', p.exists())
print('size bytes:', p.stat().st_size)
print('txt files :', [f.name for f in Path('.').glob('*.txt')])
```

## 5. CSV files with the `csv` module

```python
import csv
with open('students.csv', 'w', newline='', encoding='utf-8') as f:
    w = csv.writer(f)
    w.writerow(['name', 'cgpa'])
    w.writerow(['Asha', 8.4])
    w.writerow(['Ravi', 7.9])
    w.writerow(['Mira', 9.1])

with open('students.csv', newline='', encoding='utf-8') as f:
    for row in csv.DictReader(f):      # each row -> dict keyed by header
        print(row['name'], '->', row['cgpa'])
```

## 6. JSON files with the `json` module

```python
import json
config = {'name': 'Asha', 'subjects': ['ML', 'OS'], 'cgpa': 8.4, 'active': True}
with open('config.json', 'w', encoding='utf-8') as f:
    json.dump(config, f, indent=2)

with open('config.json', encoding='utf-8') as f:
    data = json.load(f)
print('loaded back:', data)
print('first subject:', data['subjects'][0], '| type preserved:', type(data['active']).__name__)
```

## 7. Handle file errors gracefully

```python
try:
    with open('does_not_exist.txt', encoding='utf-8') as f:
        f.read()
except FileNotFoundError as e:
    print('Handled:', e)
```

## 8. Mini-project — read a CSV, compute, write a report
A tiny data pipeline: **read → process → write**.

```python
import csv
with open('students.csv', newline='', encoding='utf-8') as f:
    students = list(csv.DictReader(f))

cgpas = [float(s['cgpa']) for s in students]
average = sum(cgpas) / len(cgpas)
topper = max(students, key=lambda s: float(s['cgpa']))

with open('report.txt', 'w', encoding='utf-8') as f:
    f.write('CLASS REPORT\n')
    f.write(f'Students : {len(students)}\n')
    f.write(f'Average  : {average:.2f}\n')
    f.write(f'Topper   : {topper["name"]} ({topper["cgpa"]})\n')

print(Path('report.txt').read_text(encoding='utf-8'))
```
