### Q1. FizzBuzz for 1..20 (clean loop / comprehension)
**Concept:** truthiness + `string * bool`. `'Fizz' * True` is `'Fizz'`; `'Fizz' * False` is `''` (empty). An empty string is falsy, so `'' or n` yields the number.

**Explain to students:** read each term as “add Fizz if divisible by 3, add Buzz if by 5; if neither, fall back to the number.”

```python
# loop version
for n in range(1, 21):
    print('Fizz'*(n % 3 == 0) + 'Buzz'*(n % 5 == 0) or n, end=' ')
print()

# one-line comprehension version
print(['Fizz'*(n%3==0) + 'Buzz'*(n%5==0) or n for n in range(1, 21)])
```

### Q2. Two numbers that sum to a target — O(n) with a dict
**Concept:** trade memory for speed. Store each value→index as you go; for each value check if its complement was already seen.

**Explain to students:** the naive solution checks every pair (O(n²)); the dictionary makes each lookup O(1), so the whole scan is O(n).

```python
def two_sum(nums, target):
    seen = {}                      # value -> index
    for i, v in enumerate(nums):
        if target - v in seen:     # complement already seen?
            return (seen[target - v], i)
        seen[v] = i
    return None

print(two_sum([2, 7, 11, 15], 9))   # (0, 1)
print(two_sum([3, 2, 4], 6))        # (1, 2)
```

### Q3. A lazy Fibonacci generator
**Concept:** `yield` makes a generator that produces values one at a time and *remembers its place* — only two numbers are ever in memory.

**Explain to students:** the `while True` loop looks infinite, but it's safe because the caller decides how many values to take (`islice`).

```python
import itertools
def fib():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

print('first 10:', list(itertools.islice(fib(), 10)))
```

### Q4. Read an int and handle non-numbers with `try/except`
**Concept:** EAFP — *try* the conversion and handle the failure, rather than checking every character first.

**Explain to students:** `int('abc')` raises `ValueError`; we catch it and respond gracefully. (In class, replace the sample list with `input()`.)

```python
def read_int(raw):
    try:
        return int(raw)
    except ValueError:
        print(f'{raw!r} is not a valid integer')
        return None

for sample in ['42', 'abc', '  7 ']:
    print(sample, '->', read_int(sample))
```

### Q5. Most common character with `collections.Counter`
**Concept:** `Counter` tallies hashable items in one line; `.most_common(k)` returns the top *k*.

**Explain to students:** prefer a ready-made tool over a manual `dict` of counts — shorter and less error-prone.

```python
from collections import Counter
s = 'banana'
counts = Counter(s)
print('counts:', counts)
print('most common:', counts.most_common(1))   # [('a', 3)]
```

### Q6. Palindrome check + time complexity
**Concept:** normalise (keep only letters/digits, lowercase) then compare the string with its reverse.

**Complexity:** **O(n)** time — cleaning, reversing (`[::-1]`) and comparing are each linear; **O(n)** extra space for the cleaned copy.

**Explain to students:** `s[::-1]` is the Pythonic reverse; `isalnum()` ignores spaces and punctuation.

```python
def is_palindrome(s):
    cleaned = ''.join(ch.lower() for ch in s if ch.isalnum())
    return cleaned == cleaned[::-1]

for t in ['Madam', 'A man, a plan, a canal: Panama', 'python']:
    print(f'{t!r:42s} -> {is_palindrome(t)}')
```
